```python
# Transition Probabilities
states=['high','low']
actions=["search", "wait", "recharge"]
transition_probabilities = {
    "high": {
        "search": {"high": 0.7, "low": 0.3},
        "wait": {"high": 1},
        "recharge": {"high": 0, "low": 0}  # Explicitly setting no transitions for recharge action
    },
    "low": {
        "search": {"high": 0.1, "low": 0.9},
        "wait": {"low": 1},
        "recharge": {"high": 1}
    }
}

# Rewards based on start state, action, and ending state
rewards = {
    "high": {
        "search": {"high": rsearch, "low": rsearch},
        "wait": {"high": rwait},
        # No reward for recharge action
        "recharge":{"high":''}
    },
    "low": {
        "search": {"high": -3, "low": rsearch},
        "wait": {"low": rwait},
        "recharge": {"high": 0}
    }
}

gamma = 0.9  # Discount factor
theta = 0.00001  # Threshold for convergence
value = {state: 1 for state in states}  # Initialize value for each state to 0

def compute_new_value(state):
    max_value = float('-inf')
    for action in actions:
        sum_values = 0
        for next_state in states:
            prob = transition_probabilities[state][action].get(next_state, 0)
            if(prob==0):
              break
            reward = rewards[state][action].get(next_state, 0)
            sum_values += prob * (reward + gamma * value[next_state])
        max_value = max(max_value, sum_values)
    return max_value

delta = float('inf')
while delta > theta:
    delta = 0
    new_value = value.copy()
    for state in states:
        v = value[state]
        new_value[state] = compute_new_value(state)
        delta = max(delta, abs(v - new_value[state]))
    value = new_value
    print(value)
```

```python
import numpy as np

# Define states, actions, rewards, and transition probabilities
states = ["high", "low"]
actions = ["search", "wait", "recharge"]

# Transition Probabilities
alpha = 0.7
beta = 0.9
gamma = 0.9

# Reward Matrix: R[state][action]
# Rewards: search while high, search while low and stay, search while low and transition to high, wait, recharge
R = {
    "high": {
        "search": 1,
        "wait": 0.01,

    },
    "low": {
         "search": 1,
        "wait": 0.01,
        "recharge": 0
    }
}

# Value Initialization
V = {
    "high": 0,
    "low": 0
}

theta = 1e-4  # Small value for checking convergence
delta = float('inf')  # Difference between updated and old value for a state

# Value Iteration
while delta > theta:
    delta = 0

    # For each state
    for state in states:
        v = V[state]  # Store the current value of the state

        # Compute expected returns for each action
        if state == "high":
            search_val = alpha * (R["high"]["search"] + gamma * V["high"]) + \
                         (1 - alpha) * (R["high"]["search"] + gamma * V["low"])
            wait_val = R["high"]["wait"] + gamma * V["high"]
            recharge_val = R["high"]["recharge"] + gamma * V["high"]  # always stays in high

            V[state] = max(search_val, wait_val, recharge_val)

        elif state == "low":
            search_val_high_transition = beta * (-3 + gamma * V["high"])
            search_val_low_stay = (1 - beta) * (R["low"]["search"] + gamma * V["low"])
            search_val = search_val_high_transition + search_val_low_stay
            wait_val = R["low"]["wait"] + gamma * V["high"]
            recharge_val = R["low"]["recharge"] + gamma * V["high"]  # always transitions to high

            V[state] = max(search_val, wait_val, recharge_val)

        # Update delta
        delta = max(delta, abs(v - V[state]))

print(V)

# Extract policy based on optimal values
policy = {}
for state in states:
    if state == "high":
        values = [
            alpha * (R["high"]["search"] + gamma * V["high"]) + \
            (1 - alpha) * (R["high"]["search"] + gamma * V["low"]),
            R["high"]["wait"] + gamma * V["high"],
            R["high"]["recharge"] + gamma * V["high"]
        ]
    else:
        values = [
            beta * (-3 + gamma * V["high"]) + (1 - beta) * (R["low"]["search"] + gamma * V["low"]),
            R["low"]["wait"] + gamma * V["high"],
            R["low"]["recharge"] + gamma * V["high"]
        ]

    policy[state] = actions[np.argmax(values)]

print(policy)
```
