```python
import numpy as np

# Define the state space, action space, and initialize the values
states = ["unemployed", "grad school", "academia", "industry"]
actions = ["Research", "D"]

# Rewards for each state
rewards = {
    "unemployed": 0,
    "grad school": 0,
    "academia": 10,
    "industry": 100
}


transition_probabilities = {
    "unemployed": {
        "Research": { "grad school": 1},
        "D": {"industry": 1},
    },
    # ... Define for all states and actions
    "grad school": {
        "Research": { "grad school": 0.1,"academia": 0.9},
        "D": {"industry": 0.9,"grad school": 0.1},
    },
    "industry": {
        "Research": { "industry": 0.1,"unemployed": 0.9},
        "D": {"industry": 1},
    },
    "academia": {
        "Research": { "academia": 1},
        "D": {"industry": 0.9,"academia": 0.1},
    },
}

gamma = 0.9  # Discount factor
theta = 100#0.00001  # Threshold for convergence
value = {state: 0 for state in states}  # Initialize value for each state to 0

def compute_new_value(state):
    max_value = float('-inf')
    i=0
    for action in actions:
        sum_values = 0
        for next_state in states:
            #i=i+1
            prob = transition_probabilities[state][action].get(next_state, 0)
            print(state,next_state,action,'prob',prob)
            reward = rewards[next_state]
            print(reward)
            sum_values += prob * (reward + gamma * value[next_state])
            print('sum_values',sum_values)
            #if(i>2):
            #  print('i',i)
              #break
        max_value = max(max_value, sum_values)
        print('max_value',max_value)
    return max_value

delta = float('inf')
while delta > theta:
    delta = 0
    new_value = value.copy()
    for state in states:
        v = value[state]
        new_value[state] = compute_new_value(state)
        delta = max(delta, abs(v - new_value[state]))
    value = new_value
    print('state',state,'v',v,'new_value',new_value,'delta',delta)
    print(value)

print(value)
```

```python
# Given: transition_probabilities, rewards, value

def compute_expected_action_value(state, action, value):
    expected_value = 0
    for next_state in states:
        trans_prob = transition_probabilities[state][action].get(next_state, 0)
        reward = rewards[next_state]
        expected_value += trans_prob * (reward + gamma * value[next_state])
    return expected_value

optimal_policy = {}

for state in states:
    best_action = None
    best_value = float('-inf')

    for action in actions:
        action_value = compute_expected_action_value(state, action, value)
        print('state',state,'action',action,action_value)
        if action_value > best_value:
            best_value = action_value
            best_action = action

    optimal_policy[state] = best_action

print(optimal_policy)
```
