```python
#import libraries
import numpy as np
import pandas as pd
```

```python
# given values and possible state & action
rsearch = 1
rwait = 1

S=["unemployed", "grad school", "academia", "industry"]
A=["R", "D"]
gamma=0.9
Teta=0.00001  # Threshold - consider a small number
#Probabilities
Action_change_Prob={"unemployed": {"R": { "grad school": 1},"D": {"industry": 1},},"grad school": {"R": { "grad school": 0.1,"academia": 0.9},"D": {"industry": 0.9,"grad school": 0.1},},
                    "industry": {"R": { "industry": 0.1,"unemployed": 0.9},"D": {"industry": 1},},"academia": {"R": { "academia": 1},"D": {"industry": 0.9,"academia": 0.1},},}
#Rewards
Rewards={"unemployed": 0,"industry": 100,"grad school": 0,"academia": 10}

State_values={}

#pass orbitatory initial values to State-Values
for s in S:
  State_values[s]=1
# State_values['high-battery']=1
# State_values['low-battery']=1
```

```python
print('initial_state_values',State_values)
```

```python
# Action_change_Prob[s][action].get(next_state)
```

```python
d= 10000000.0 #initialized delta with a big number

while d > Teta:
    State_values_holder = State_values.copy()
    d = 0
    for s in S:
        optimal_value = float('-inf')
        #finding optimal V.
        for action in A:
          #find expected return when the agent takes an action
          expected_return = 0
          for next_state in S:
              p = Action_change_Prob[s][action].get(next_state) if (Action_change_Prob[s][action].get(next_state) is not None) else 0
              expected_return = expected_return + p * (Rewards[next_state] + gamma * State_values[next_state])
          if(expected_return>optimal_value): optimal_value=expected_return
          State_values_holder[s]=optimal_value
        d = max(d, abs(State_values[s] - State_values_holder[s]))
    State_values = State_values_holder
    print(State_values)
```
