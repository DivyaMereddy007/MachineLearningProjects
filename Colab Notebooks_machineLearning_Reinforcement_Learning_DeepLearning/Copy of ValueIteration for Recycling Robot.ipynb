```python
# given values and possible state & action
rsearch = 1
rwait = 0.0001

S=['high-battery','low-battery']
A=["search", "wait", "recharge"]
gamma=0.9
Teta=0.00001  # Threshold - consider a small number
#Probabilities
Action_change_Prob={ "high-battery": {"search": {"high-battery": 0.7, "low-battery": 0.3},"wait": {"high-battery": 1},"recharge": {"high-battery": 0, "low-battery": 0},},
    "low-battery": {"search": {"high-battery": 0.1, "low-battery": 0.9},"wait": {"low-battery": 1},"recharge": {"high-battery": 1}}
                     }
#Rewards
Rewards={"high-battery": {"search": {"high-battery": rsearch, "low-battery": rsearch},"wait": {"high-battery": rwait},"recharge":{"high-battery":None}},
    "low-battery": {"search": {"high-battery": -3, "low-battery": rsearch},"wait": {"low-battery": rwait},"recharge": {"high-battery": 0}}}

State_values={}

#pass orbitatory initial values to State-Values
State_values['high-battery']=1
State_values['low-battery']=1
```

```python
print('initial values of Value functions',State_values)
```

```python
d= 10000000.0 #initialized delta with a big number

while d > Teta:
    State_values_holder = State_values.copy()
    d = 0
    for s in S:
        optimal_value = float('-inf')
        #finding  state action value functions
        for action in A:
          expected_return = 0
          for next_state in S:
              p = Action_change_Prob[s][action].get(next_state) if (Action_change_Prob[s][action].get(next_state) is not None) else 0
              if(p==0):
                break
              r = Rewards[s][action].get(next_state) if (Rewards[s][action].get(next_state) is not None) else 0
              expected_return = expected_return+ p * (r + gamma * State_values[next_state])
          if(expected_return>optimal_value): optimal_value=expected_return
          State_values_holder[s]=optimal_value
        d = max(d, abs(State_values[s] - State_values_holder[s]))
    State_values = State_values_holder
    print(State_values)
```
